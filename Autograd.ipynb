```python
import math
import torch
import numpy as np
import matplotlib.pyplot as plt
```

## y = 3x - 6x^3 - x^4 + 4

```python
itr_count = 0
min_max_list = []
y_list = []
x_list = []

# update_rate = 5e-4
threshold = 4e-3

#setting lower limit for x
x = torch.tensor(-6.0, requires_grad=True) 


def step_decay(epoch):
   initial_lrate = 0.01
   drop = 0.5
   epochs_drop = 100.0
   lrate = initial_lrate * math.pow(drop,  
           math.floor((1+epoch)/epochs_drop))
   return lrate


#while x less than equal to upper limit of x
while x<=1:
    update_rate = step_decay(itr_count)
#     print(update_rate)
    #define the equation for y
    y = 3*x - 6*x**3 - x**4 + 4
    
    #appending values of x and y to x_list & y_list
    y_list.append(y.data)
    x_list.append(x.data)
    
    #finding derivatives of y with respect to x
    y.backward()
    
    #finding the slope
    slope = x.grad
#     print(slope)
    #compare the slope with threshold and append to list if it satisfies the condition of slope<=threshold
    if abs(slope)<=threshold:

        min_max_list.append((x.data,y.data))

    if itr_count <= 1000:
       x.data = x.data + update_rate
    elif itr_count <= 10000:
       x.data = x.data + update_rate
    else:
       x.data = x.data + (abs(x.grad)*update_rate)
        
    #Update the iteartion count
    itr_count+=1
    
    #clear the gradients
    x.grad.zero_()   
  
# print("min_max_list: ", min_max_list)
#plot the contour and mark local maxima and minima
# fig = plt.figure()
# ax = fig.add_subplot(111)
# fig.suptitle('Optimization Contour', fontsize='large')
# ax.set_xlabel('x', fontsize='medium')
# ax.set_ylabel('y', fontsize='medium')

# x_values = [float(A.numpy()) for A in x_list]
# y_values = [float(A.numpy()) for A in y_list]

 

# _ = ax.plot(x_values, y_values, c='k')


# for i in range(len(min_max_list)):
#     ax.scatter(min_max_list[i][0], min_max_list[i][1], c='r', marker='D')
```

 ## Z = cos(x)*sin(y)

```python
torch.pi = torch.acos(torch.zeros(1)).item() * 2
```

```python
x = torch.tensor(-torch.pi, requires_grad=True)
y = torch.tensor(-torch.pi, requires_grad=True)
z = torch.cos(x)*torch.sin(y)

```

```python
count = 0
min_max_list = []
y_list = []
x_list = []
z_list = []

update_rate = 9e-3
threshold = 7e-3

#setting lower limit for x and y
x = torch.tensor(-torch.pi, requires_grad=True)
y = torch.tensor(-torch.pi, requires_grad=True)

#while x less than equal to upper limit of x and y less than equal to upper limit of y
while x<=torch.pi and  y <=torch.pi:
    
    #define the equation for z
    z = torch.cos(x)*torch.sin(y)
    
    #appending values of x and y to x_list & y_list
    y_list.append(y.data)
    x_list.append(x.data)
    z_list.append(z.data)
      
    #finding derivatives of y with respect to x
    z.backward()
    
    #finding the slope
    slope_x= x.grad
    slope_y = y.grad
    
    slope=slope_x+slope_y

    #compare the slope with threshold and append to list if it satisfies the condition of slope<=threshold
    if abs(slope)<=threshold:

        min_max_list.append((x.data,y.data,z.data))
    
    # Custom Update Schedule
    #Gradient value can be high for intial iteration which may result in large value of x. 
    #So for few intial iterations gradient value is ignored
    if count < 3:
        x.data = x.data + update_rate
        y.data = y.data + update_rate
    else:
        x.data = x.data + (abs(x.grad)*update_rate)
        y.data = y.data + (abs(y.grad)*update_rate)
    
    #clear the gradients
    x.grad.zero_()
    y.grad.zero_()
    
print("min_max_list: ", min_max_list)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlabel('x', fontsize='medium')
ax.set_ylabel('y', fontsize='medium')   
ax.set_zlabel('z', fontsize='medium') 

x_values = [float(A.numpy()) for A in x_list]
y_values = [float(A.numpy()) for A in y_list]
z_values = [float(A.numpy()) for A in z_list]

ax.plot3D(x_values, y_values, z_values, 'gray')


for i in range(len(min_max_list)):
    ax.scatter3D(min_max_list[i][0], min_max_list[i][1], min_max_list[i][2], c='r', marker='D')
```
